```python
import pandas as pd
import numpy as np
import re
import pickle
```

```python
file_path = '../data/'
def pklopen(fine_name: str):
    with open(file_path + fine_name, 'rb') as f:
        return pickle.load(f)
```

```python
df1 = pklopen('naverCafe_crawling(우크라이나, 20230101-20231231).pkl')
df2 = pklopen('daumCafe_crawling(우크라이나, 20230101-20231231).pkl')

print(df1)
print(df2)
```

```python
# 데이터 결합
data = pd.concat([df1, df2])
data
```

```python
# 그룹별로 사이즈 확인
data.groupby(['ch1', 'ch2']).size()
```

```python
data = data.reset_index(drop=True)  # 인덱스 초기화
data
```

```python
# 데이터 pickle로 저장
file_path = '../data/'
file_name = 'total_data.pkl'
data.to_pickle(file_path + file_name)
```

```python
with open(file_path + file_name, 'rb') as f:
    docs = pickle.load(f)
docs
```

```python
arr = ['1', '2', '3']
print(', '.join(arr))
```

```python
docs['comment'] = docs['comment'].apply(lambda x: ', '.join(x) if isinstance(x, list) else x)
docs['comment']
```

```python
# 제목, 본문. 댓글 한글화
for i in range(len(docs)):
    docs['title'][i] = re.sub('[^가-힣]', ' ', docs['title'][i])
    if isinstance(docs['main_text'][i], str):
        docs['main_text'][i] = re.sub('[^가-힣]', ' ', docs['main_text'][i])
    if isinstance(docs['comment'][i], str):
        docs['comment'][i] = re.sub('[^가-힣]', ' ', docs['comment'][i])
docs
```
