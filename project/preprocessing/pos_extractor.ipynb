```python
import pandas as pd
from konlpy.tag import Mecab
import multiprocessing as mp
import parmap
```

```python
mecab = Mecab(dicpath='C://mecab//mecab-ko-dic')
```

```python
df = pd.read_feather("../data/posTagged_total.feather", use_threads=True)
df
```

```python
# 추출할 품사 리스트
desired_pos = ['NNG', 'NNP', 'VV', 'VA', 'MAG', 'NR', 'MM']     # 일반명사, 고유명사, 동사, 형용사, 일반부사, 수사, 관형사
```

```python
# 추출된 품사 필터링
def filtering(pos_list: list):
    for pos in pos_list:
        if pos[1] in desired_pos:
            return pos[0]
```

```python
df["Tokenized"] = parmap.map(filtering, df["Pos"], pm_pbar=True, pm_processes=mp.cpu_count())
```
