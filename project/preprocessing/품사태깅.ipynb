```python
# Mecab 객체 생성
mecab = Mecab(dicpath='C://mecab//mecab-ko-dic')

# 품사 태깅 함수
def pos_tagging(text):
    return mecab.pos(text)

# 명사, 형용사, 동사만 추출하는 함수
def filter_pos(tagged):
    # 사용할 품사 태그들
    allowed_pos = {'NNG', 'NNP', 'VA', 'VV'}
    return [word for word, pos in tagged if pos in allowed_pos]
```

```python
# "Tokenized" 열의 단어들에 대해 품사 태깅 수행 후, 명사/형용사/동사만 추출
docs['Filtered_Pos'] = docs['Tokenized'].apply(lambda x: filter_pos(pos_tagging(' '.join(x))))

# 결과 확인
print(docs.head())
```
